```python
import torch
from transformers import AutoProcessor
```

```python
tokenizer = AutoProcessor.from_pretrained("unsloth/csm-1b")
```

```python
import torch

def preprocess(example):
  conversation=[
      {
          'role':'0',
          'content':[
              {'type':'text','text':example['text']},
              {'type':'audio','path':example['audio']['array']}
          ],
      }

  ]
  try:
    # Ensure max_length and max_audio_length are defined and accessible
    global max_length, max_audio_length

    model_inputs=tokenizer.apply_chat_template(
        conversation,
        tokenize=True,
        return_dict=True,
        output_labels=True,
        text_kwargs={
            'padding':'max_length',
            'max_length':max_length+12,
            'pad_to_multiple_of':8,
            'padding_side':'right',
        },
        audio_kwargs={
            "sampling_rate": 24000,
            "max_length": max_audio_length+12, # max input_values length of the whole dataset
            "padding": "max_length",
        },
        common_kwargs={
            'return_tensors':'pt'
        }
    )
  except Exception as e:
    print(e)
    return None

  # Ensure the output is a dictionary with tensors
  processed_example = {}
  for key, value in model_inputs.items():
      if isinstance(value, torch.Tensor):
          # Squeeze the tensor to remove the batch dimension if it exists
          processed_example[key] = value.squeeze(0)
      else:
          # Handle potential non-tensor outputs if necessary
          processed_example[key] = value


  # Ensure all required keys are present and are tensors
  required_keys = ["input_ids", "attention_mask", "labels", "input_values", "input_values_cutoffs"]
  for key in required_keys:
      if key not in processed_example or not isinstance(processed_example[key], torch.Tensor):
           print(f"Error: Required key '{key}' is missing or not a tensor in processed example.")
           return None


  return processed_example
```

```python
ds=ds.map(
    preprocess,
    remove_columns=ds.column_names,
    desc='Preprocessing dataset'
)



```

```python
import sklearn
```

```python
from sklearn.model_selection import train_test_split

train_ds,test_ds=ds.train_test_split(test_size=0.2,shuffle=True,seed=42)
```

```python
from transformers import TrainingArguments, Trainer
from unsloth import is_bfloat16_supported

trainer = Trainer(
    model = model,
    train_dataset = ds,

    args=TrainingArguments(
        per_device_train_batch_size=2,
        gradient_accumulation_steps=8,
        warmup_steps=5,
        max_steps=100,
        learning_rate=2e-4,
        fp16=not is_bfloat16_supported(),
        bf16=is_bfloat16_supported(),
        logging_steps=1,
        optim='adamw_8bit',
        weight_decay=0.01,
        lr_scheduler_type='linear',
        seed=3400,
        output_dir='outputs',
        report_to='none',
        logging_dir=f'logs/{new_model_name}',
        gradient_checkpointing=True,
    ),
)
```

```python
training=trainer.train()
```
